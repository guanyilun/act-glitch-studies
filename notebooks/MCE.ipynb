```python
import moby2
from moby2.scripting import get_filebase
import cPickle
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
fb = get_filebase()
```

```python
cut_id = 3333
cut_data = cPickle.load(open("../outputs/cuts/%d.cut" % cut_id, "r"))
```

```python
cut_data.keys()
```

```python
cut_data['mce'].cuts
```

```python
tod_name = cut_data['TOD']
tod_dir = fb.filename_from_name(tod_name, single=True)
tod_data = moby2.scripting.get_tod({'filename':tod_dir, 'repair_pointing':True})    
```

```python
tod_data.data
```

```python
pixel_id = 190
```

```python
print cut_data['cuts'].cuts[pixel_id]
print cut_data['mce'].cuts[pixel_id]
plt.plot(tod_data.data[pixel_id])
```
