# Energy per Detector

```python
data = np.genfromtxt('0_100_tods_hist.txt')
plt.figure(figsize=(12,7))
data = np.log(data)
        
slope, intercept = np.polyfit(data[0],data[1],1)
data_y = slope* data[0] + intercept
plt.plot(data[0],data_y, label='Slope = ' + str(slope))

plt.legend()
plt.step(*data)
plt.xlabel('$log$(Energy) $pJ$')
plt.ylabel('$log$(Events)')
plt.title('Energy per (Covered) Detector 0-100 TODs')
plt.show()
```

```python
data_unf = np.genfromtxt('0_150_unfiltered_tods_hist.txt')
data_f = np.genfromtxt('0_150_filtered_tods_hist.txt')
data_unc = np.genfromtxt('10000_10150_unf_tods_hist.txt')
data_func = np.genfromtxt('10000_10150_crf_tods_hist.txt')


data_unf = np.log(data_unf)
data_f = np.log(data_f)
data_unc = np.log(data_unc)
data_func = np.log(data_func)

plt.figure(figsize=(15,10))

plt.step(*data_unf, label="Unifiltered, Covered")
plt.step(*data_f,label="CR-Filtered, Covered")
plt.step(*data_unc,label="Unfiltered, Uncovered")
plt.step(*data_func,label="Filtered, Uncovered")
plt.xlabel('$log$(Energy) $pJ$')
plt.ylabel('$log$(Events)')
plt.title('Energy per Detector 0-150 TODs')
plt.legend()
plt.show()
```

# Energy per Event

```python
data1 = np.genfromtxt('0_150_unf_events_hist.txt') #Unfiltered, covered
data2 = np.genfromtxt('0_150_crf_events_hist.txt') #Filtered, covered
data3 = np.genfromtxt('10000_10150_unf_events_hist.txt') #Unfiltered, uncovered
data4 = np.genfromtxt('10000_10150_crf_events_hist.txt') #Filtered, uncovered


data1 = np.log(data1)
data2 = np.log(data2)
data3 = np.log(data3)
data4 = np.log(data4)

plt.figure(figsize=(15,10))

plt.step(*data1, label="Unifiltered, Covered")
plt.step(*data2,label="CR-Filtered, Covered")
plt.step(*data3,label="Unfiltered, Uncovered")
plt.step(*data4,label="CR-Filtered, Uncovered")
plt.xlabel('$log$(Energy) $pJ$')
plt.ylabel('$log$(Events)')
plt.title('Energy per Event 0-150 TODs')
plt.legend()
plt.show()
```

```python
slope1, intercept1 = np.polyfit(data1[0][2:22],data1[1][2:22],1)
data1y = slope1*data1[0][2:22] + intercept1

slope2, intercept2 = np.polyfit(data2[0][2:22],data2[1][2:22],1)
data2y = slope2*data2[0][2:22] + intercept2



plt.figure(figsize=(12,7))
plt.plot(data1[0][2:22],data1y,'--', label='Slope,unf = ' + str(slope1))
plt.plot(data2[0][2:22],data2y,'--', label='Slope,filt = ' + str(slope2))



plt.step(*data1, label="Unfiltered")
plt.step(*data2,label="CR-Filtered")
plt.xlabel('$log$(Energy) $pJ$')
plt.ylabel('$log$(Events)')
plt.title('Energy per Event, 150 Covered TODs')
plt.legend()
plt.show()
```
