```python
%pylab inline
from scipy.interpolate import interp1d
```

```python
data = np.genfromtxt('0_100_tods_hist.txt')
plt.figure(figsize=(12,7))
data = np.log(data)
        
slope, intercept = np.polyfit(data[0],data[1],1)
data_y = slope*data[0] + intercept
plt.plot(data[0],data_y, label='Slope = ' + str(slope))

plt.legend()
plt.xscale('log')
plt.yscale('log')
plt.step(*data)
plt.xlabel('$log$(Energy) $pJ$')
plt.ylabel('$log$(Events)')
plt.title('Energy per (Covered) Detector 0-100 TODs')
plt.show()
```

```python
data_unf = np.genfromtxt('0_150_unfiltered_tods_hist.txt')
data_f = np.genfromtxt('0_150_filtered_tods_hist.txt')


data_unf = np.log(data_unf)
data_f = np.log(data_f)
        
slope_unf, intercept_unf = np.polyfit(data_unf[0],data_unf[1],1)
data_unfy = slope_unf*data_unf[0] + intercept_unf

slope_f, intercept_f = np.polyfit(data_f[0],data_f[1],1)
data_fy = slope_f*data_f[0] + intercept_f

plt.figure(figsize=(15,10))
#plt.plot(data_unf[0],data_unfy, label='Slope = ' + str(slope_unf))
#plt.plot(data_f[0],data_fy, label='Slope = ' + str(slope_f))

plt.xscale('log')
plt.yscale('log')
plt.step(*data_unf, label="Unifiltered")
plt.step(*data_f,label="CR-Filtered")
plt.xlabel('$log$(Energy) $pJ$')
plt.ylabel('$log$(Events)')
plt.title('Energy per (Covered) Detector 0-150 TODs')
plt.legend()
plt.show()
```
